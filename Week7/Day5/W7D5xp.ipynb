```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pprint  # To tidy up

options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run Chrome in headless mode
options.add_argument("--no-sandbox")  # Bypass OS security model
options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems
driver = webdriver.Chrome(options=options)



url = "https://www.inmotionhosting.com"
driver.get(url)

wait = WebDriverWait(driver, 30)
plan_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'compare-1')))

child_elements = []

for parent_element in plan_elements:
    child_elements = parent_element.find_elements(By.CLASS_NAME, 'imh-rostrum-card')

child_values = [c.text for c in child_elements]
child_values

plan_dict = {}
plan_names = []
descriptions = []
prices = []


for i, p in enumerate(child_values):
    p = p.split('\n')
    if p[0] == 'NEW':
        plan_names.append(p[1])
        descriptions.append(p[2])
        prices.append(p[4])
        plan_dict['plan_name'] = plan_names
        plan_dict['description'] = descriptions
        plan_dict['price'] = prices
    else:
        plan_names.append(p[0])
        descriptions.append(p[1])
        prices.append(p[3])
        plan_dict['plan_name'] = plan_names
        plan_dict['description'] = descriptions
        plan_dict['price'] = prices

plan_dict

import pandas as pd

df = pd.DataFrame(plan_dict)

df
```
