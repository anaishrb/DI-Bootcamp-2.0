```python
#Exercise1
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

np.random.seed(42)

num_data_points = 1000

A = np.random.randint(0, 1001, size=num_data_points)
B = np.random.uniform(0, 10, size=num_data_points)

data = pd.DataFrame({
    'FeatureA': A,
    'FeatureB': B
})

print(data.head())

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)

print("Normalized Data:")
print(df_normalized)

plt.figure(figsize=(8, 6))
plt.scatter(data['FeatureA'], data['FeatureB'], alpha=0.5)
plt.title('Scaled Dataset Scatter Plot')
plt.xlabel('FeatureA (Scaled)')
plt.ylabel('FeatureB (Scaled)')
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_standardized = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)

print("Standardized Data:")
print(df_standardized)
plt.figure(figsize=(8, 6))
plt.scatter(data['FeatureA'], data['FeatureB'], alpha=0.5)
plt.title('Normalized Dataset Scatter Plot')
plt.xlabel('FeatureA (Normalized)')
plt.ylabel('FeatureB (Normalized)')
plt.grid(True)
plt.show()
```

Min-Max Scaling in this dataset preserve the original range of FeatureA (0 to 1000) and FeatureB (0 to 10). This method is better to preserve the original data range.
