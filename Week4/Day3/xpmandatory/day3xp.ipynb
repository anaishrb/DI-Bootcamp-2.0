```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 28, 35, 27],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
}
df = pd.DataFrame(data)

excel_filename = 'dataframe_export.xlsx'
df.to_excel(excel_filename, index=False)
print(f"DataFrame exported to '{excel_filename}' (Excel file).")

json_filename = 'dataframe_export.json'
df.to_json(json_filename, orient='records')
print(f"DataFrame exported to '{json_filename}' (JSON file).")
```
