```python
#EXERCISE 4
data = 'sales_data.csv'
data = pd.read_csv(data)

total_quantity_sold = data['quantity'].sum()
print(f"Total quantity of products sold: {total_quantity_sold}")

category_revenue = data.groupby('category')['revenue'].sum()
highest_revenue_category = category_revenue.idxmax()
highest_revenue = category_revenue.max()
print(f"Category with the highest revenue: {highest_revenue_category}")
print(f"Revenue generated by the highest revenue category: ${highest_revenue:.2f}")
average_revenue_per_sale = data['revenue'].mean()
print(f"Average revenue per sale: ${average_revenue_per_sale:.2f}")


```
