```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.load_dataset('iris')

df = sns.load_dataset('iris')

from plotnine import ggplot, aes, geom_point
import seaborn as sns
print(df)
```

```python
scatter_plot = (
    ggplot(df, aes(x='sepal_length', y='sepal_width', color='species')) +
    geom_point()
)

#I did not manage to write a title or the name of the axes
print(scatter_plot)
```

```python
import numpy as np
plt.hist(df['petal_length'], bins=35, edgecolor='black')
plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#Seaborn Task: Produce a pair plot to show pairwise relationships in the dataset.
# Create a pair plot
sns.pairplot(df, hue='species')
plt.title('Pair Plot of Iris Dataset', y=1.02)
plt.show()

#I couldn't place the title in the right position...
```

```python
#Plotly Task: Create an interactive 3D scatter plot mapping sepal length, sepal width, and petal length, color-coded by species.
import plotly.express as px
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_length', color='species')
fig.show()

fig.update_layout(
    title='Customized Plot',
    xaxis_title='Sepal Length',
    yaxis_title='Sepal Width',
    scene=dict(zaxis_title='Petal Length'),
    template='plotly_dark'
)
fig.show()
```
