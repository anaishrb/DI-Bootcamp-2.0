```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ex2data1.txt", header=None, names=['Exam1', 'Exam2', 'Admission'])

print(df.head())

plt.scatter(df[df['Admission'] == 1]['Exam1'], df[df['Admission'] == 1]['Exam2'], label='Admitted', marker='o')
plt.scatter(df[df['Admission'] == 0]['Exam1'], df[df['Admission'] == 0]['Exam2'], label='Not Admitted', marker='x')

plt.xlabel('Exam 1 Score')
plt.ylabel('Exam 2 Score')
plt.legend()

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X = df[['Exam1', 'Exam2']]
y = df['Admission']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```

```python
all_predictions = model.predict(X)

# Calculate accuracy on the entire dataset
overall_accuracy = accuracy_score(y, all_predictions)

# Print the accuracy
print(f"Overall Accuracy: {overall_accuracy:.2f}")
```

An  accuracy of 0.89 (or 89%) suggests that the logistic regression model is performing well on the dataset. This accuracy metric indicates the proportion of correctly predicted outcomes (both admitted and not admitted) over the entire dataset.
